# Packages et modules en Python

## Créer un module

* Chaque fichier '.py' est un module
* Chaque répertoire qui contient un fichier `__init__.py` est un package

* Répertoire de recherche :

~~~
import sys
sys.path # liste que l'on peut modifier
~~~

* Attention, un module n'est chargé qu'une seule foi !

* Import absolu et relatif : dans un package, on peut importer un module en utilisant un chemin relatif 
~~~
from .my_packate import another_module
~~~

## Installer un module 

Lorsque l'on veut utiliser un moduler pour tout le système, il est possible de l'installer globalement. Pour cela on écrit un fichier `setup.py` qui permettra de faire pluseurs opérations : installer le module, ziper le module et même le déposer sur un serveur.


# Tester un module

Il faut toujours tester son code au fur et à mesure. Faire des tests très simples (99% des erreurs sont simples). Faire le plus de tests possible. 

Pacakge `unittest` : permet de distribuer les tests et de les automatiser. 
~~~
import unittest

class MyTest(unittest.TestCase):
    def test_somme(self):
        self.assertEqual(1+1, 2)
~~~

Il existe plusieurs assert (Equal, True, Less, ...). Regarder https://docs.python.org/3/library/unittest.html


* Toujours utiliser des tests lors d'optimisation du code

# Utilisation de git

Git est un logiciel de gestion de source

* Permet de suivre les modifications
* Permet de travailler à plusieurs sur un même projet
* Permet de partager facilement son travail

Utilisation en ligne de commande 
~~~ 
git init
git add fichier directory/
git commit -m 'intitial commit'
git checkout -b 'new_branch'
git checkout main
~~~
